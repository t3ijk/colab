```python
import torch 

a = torch.randn((3,3), requires_grad = True)

w1 = torch.randn((3,3), requires_grad = True)
w2 = torch.randn((3,3), requires_grad = True)
w3 = torch.randn((3,3), requires_grad = True)
w4 = torch.randn((3,3), requires_grad = True)

b = w1*a 
c = w2*a

d = w3*b + w4*c 

L = 10 - d

leafs = tuple(map(lambda t: t.is_leaf, (a, b, c, d, L)))
gradfns = tuple(map(lambda t: t.grad_fn, (a, b, c, d, L)))
shapes = tuple(map(lambda t: t.shape, (a, b, c, d, L)))

# 参考：https://blog.paperspace.com/pytorch-101-understanding-graphs-and-automatic-differentiation/	
# 将打印是否时叶子，是否有 gradfns
# 在这里叶子一般是用户直接输入的 tensor 而不是计算生成的 tensor。 个人理解用户直接输入的 tensor 是数据，然经过计算的 tensor 其实是 intermediate-tensor 数据或者这样的tensor本质是 operate 数据只是输入'形参' '形tensor'

print(leafs, '\n', gradfns, '\n', shapes)
```

```python
# RuntimeError: grad can be implicitly created only for scalar outputs
# backward 只能处理输出是 scale 的函数， 然默认 tensor 是 vector
L.backward()

```

```python
# 将 L 改为 scale 的输出
L = (10 - d).sum()
print(L.is_leaf, L.grad_fn, L.shape)
L.backward()
```
