```python
import errant

annotator = errant.load('en')
# orig = annotator.parse('This are gramamtical sentence .')
# cor = annotator.parse('This is a grammatical sentence .')
orig = annotator.parse('To hight trainning efficient each minibatch constructed by grouping sentences of similar length together.')
cor = annotator.parse('To make training more efficient, each minibatch is constructed by grouping sentences of similar length.')
edits = annotator.annotate(orig, cor)
for e in edits:
    print(e.o_start, e.o_end, e.o_str, e.c_start, e.c_end, e.c_str, e.type)
```
